```python
from obspy import Trace
import cartopy.crs as ccrs
import pylab as plt
import numpy as np

# test that cartopy works
fig, ax = plt.subplots(1, 1, subplot_kw={"projection": ccrs.Orthographic()})
ax.coastlines()
plt.show(fig)

# test that obspy works
tr = Trace()
tr.data = np.random.uniform(low=-1, high=1, size=1000)
fig, ax = plt.subplots(1, 1)
ax.plot(tr)
plt.show(fig)
```
